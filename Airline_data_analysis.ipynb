# The Flights dataset

Here, the the Flight booking dataset of various Airlines is scraped datewise in a structured format. The dataset contains the records of flight travel details between cities in India. Here ,multiple features are present like Source & Destination city, Arrival & departure time, Duration & Price of the flight, etc.

This data is available as a CSV file. We are going to analyze this dataset using the Pandas Dataframe. 

This analysis will provide valuable insights in the field of Travelling.
